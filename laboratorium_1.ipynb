```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

f = lambda x: x**2 +5
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-6, 6, 100)
x3 = np.linspace(0, 5, 100)

y1 = f(x1)
y2 = f(x2)
y3 = f(x3)

plt.plot(x1, y1)
plt.title('przedziaÅ 1')
plt.legend()
plt.xlabel('x1')
plt.ylabel('f(x1)')
plt.show()
```
